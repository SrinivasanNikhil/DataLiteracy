# Python Pandas

Pandas is an open-source library that is built on top of NumPy library. It is a Python package that offers various data structures and operations for manipulating numerical data and time series. It is mainly popular for importing and analyzing data much easier. Pandas is fast and it has high-performance & productivity for users.

Advantages of pandas
1. Fast and efficient for manipulating and analyzing data.
Data from different file objects can be loaded.
2. Easy handling of missing data (represented as NaN) in floating point as well as non-floating point data
3. Size mutability: columns can be inserted and deleted from DataFrame and higher dimensional objects
4. Data set merging and joining.
Flexible reshaping and pivoting of data sets
5. Provides time-series functionality.
6. Powerful group by functionality for performing split-apply-combine operations on data sets.

## Python Objects

Pandas generally provide two data structure for manipulating data, They are:

* Series
* DataFrame

### Series

Pandas Series is a one-dimensional labeled array capable of holding data of any type. The axis labels are  called index. Pandas Series is equivalent to a column in an excel sheet.

Pandas Series can be created from the lists, dictionaryand other data structures but in practice will probably be loaded from CSV or other types of files. 

A series can also be created with index values that are customized.


```python
import numpy as np
import pandas as pd

empSer = pd.Series(dtype=int) #creates an empty series

aList = [9,8,7,6,5] #creating a list

anArr = np.array([9,8,7,6,5]) #creating an array

ser1 = pd.Series(anArr) #arraysa, lists and other data structures can be made into series

print(ser1)

ser2 = pd.Series(anArr, index = ['a','b','c','d','e'])

print(ser2)
```

### DataFrame

Pandas DataFrame is two-dimensional changable, multi column, data structure with potentially heterogeneous tabular data structure with labeled axes (rows and columns). 

A Data frame is a two-dimensional data structure, i.e., data is aligned in a tabular fashion in rows and columns. 

A dataframe consists of three parts, the data, rows, and columns.


```python

import pandas as pd
 
# list of strings
aList = ['Tomatoes', 'is','considered', 'both', 'fruit', 
            'and', 'vegetable']
 
# Calling DataFrame constructor on list
df = pd.DataFrame(aList)
print(df)
print()

df = df.rename(columns={0:'Word'})

print(df)
print()
#Dictionary data type, has a structure of key:value
#keys are always unique
students = {
    'Name':['Nikhil','Akhil','Sandy'], #key is Name, value is the list of names
    'GPA':[3.2,3.3,3.4] #key is GPA, value is the list of GPAs
}

print(type(students))
print(students)
print()

studentsdf = pd.DataFrame(students)
print(studentsdf)
```

### Rows and Columns in the data frame

There are many operations that can be performed on the rows and columns of a data frame. Some suc operations are illustrated below



```python
#Selecting columns

import pandas as pd

data = pd.read_csv("https://elasticbeanstalk-us-east-1-712866102994.s3.amazonaws.com/data/Cars.csv")

data[["Car","MPG"]]
```

```python
#Not Seleccting certain columns

import pandas as pd

data = pd.read_csv("https://elasticbeanstalk-us-east-1-712866102994.s3.amazonaws.com/data/Cars.csv")

data.drop("MPG",axis=1)
```

```python
# Adding a new column

import pandas as pd

data = pd.read_csv("https://elasticbeanstalk-us-east-1-712866102994.s3.amazonaws.com/data/Cars.csv")

data["Weight(pounds)"] = data["Weight"]*2.20

data.head()
```

```python
# Deleting a column

import pandas as pd

data = pd.read_csv("https://elasticbeanstalk-us-east-1-712866102994.s3.amazonaws.com/data/Cars.csv")

data["Weight(pounds)"] = data["Weight"]*2.20

print(data)

data.drop(["Weight(pounds)"],axis = 1, inplace=True)

print(data)
```

```python
#Selecting rows

#Basic
import pandas as pd

data = pd.read_csv("https://elasticbeanstalk-us-east-1-712866102994.s3.amazonaws.com/data/Cars.csv")

rslt = data[data["Origin"] == "US"]

rslt
```

```python
import pandas as pd

data = pd.read_csv("https://elasticbeanstalk-us-east-1-712866102994.s3.amazonaws.com/data/Cars.csv")

#Using loc, also multiple conditions

rslt = data.loc[(data["MPG"] > 20) & (data['Origin'] == "US")]

rslt
```
